```python
import requests
```

```python
def fetch_api(endpoint):
    try:
        response = requests.get(f"https://verkehr.autobahn.de/o/autobahn{endpoint}")
        if response.status_code != 200:
            raise Exception
        return response.json()
    
    except Exception as e:
        ...
```

```python
details = []
```

```python
data = fetch_api("/")
roads = data["roads"]
for road in roads:
    data = fetch_api(f"/{road}/services/electric_charging_station")
    stations = data["electric_charging_station"]
    for station in stations:
        station_id = station["identifier"]
        detail = fetch_api(f"/details/electric_charging_station/{station_id}")
        details.append(detail)
```

```python
for detail in details:
    print(detail['title'])
```

```python
keys = set()

for detail in details:
    # Skip this detail if any value is None
    for k, v in detail.items():
        if v is not None:
            keys.add(f"{k}: {type(v)}")
```

```python
for detail in details:
    if t := detail["startTimestamp"] is not None:
        print(t)
```
